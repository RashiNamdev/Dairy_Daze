```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

# Load Microsoft's DialoGPT (perfect for customer support chat)
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Add padding token
tokenizer.pad_token = tokenizer.eos_token

# Create chat pipeline
chatbot = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.7,
    pad_token_id=tokenizer.eos_token_id
)

print("ðŸ¤– DialoGPT loaded! Ready for customer support chat.")
```

```python
import time
import re
from datetime import datetime

class PerfectCustomerBot:
    def __init__(self):
        self.chat_history = []
        self.user_name = "Customer"
        self.current_order = None

        # Mock database
        self.ORDERS = {
            "ORD1001": {"status": "Delivered âœ…", "date": "2025-12-20", "location": "Delhi"},
            "ORD1002": {"status": "In Transit ðŸ“¦", "date": "2025-12-28", "location": "Mumbai"},
            "ORD1003": {"status": "Pending Payment âš ï¸", "date": "2025-12-29", "location": "Bangalore"}
        }

        # Professional responses
        self.RESPONSES = {
            "greeting": ["Hello! Welcome to BlueCart Support. How can I help?",
                        "Hi there! What can I assist you with?",
                        "Hello! I'm here for all your BlueCart needs!"],
            "order_status": "Please share your order ID (like ORD1001) to check status.",
            "return": "âœ… 30-day return policy. Visit returns.bluecart.com with your order ID.",
            "shipping": "ðŸšš Free shipping over â‚¹500. Standard: 3-5 days, Express: 1-2 days.",
            "hours": "ðŸ•’ Support: Mon-Fri 9AM-6PM IST. Weekend chat available.",
            "payment": "ðŸ’³ UPI, Cards, Net Banking, COD. All secure with SSL.",
            "unknown": ["I can help with orders, returns, shipping, or hours. What specifically?",
                       "Common topics: order status, returns, shipping. Try one?",
                       "Let me assist with your order or policy questions!"]
        }

    def detect_intent(self, text):
        text = text.lower()
        if any(word in text for word in ["hi", "hello", "hey", "how are"]):
            return "greeting"
        elif any(word in text for word in ["order", "track", "status", "delivery"]):
            return "order_status"
        elif any(word in text for word in ["return", "refund"]):
            return "return"
        elif any(word in text for word in ["ship", "delivery time"]):
            return "shipping"
        elif any(word in text for word in ["hour", "time", "open"]):
            return "hours"
        elif any(word in text for word in ["pay", "payment"]):
            return "payment"
        return "unknown"

    def extract_order_id(self, text):
        match = re.search(r'ORD\d{4}', text.upper())
        return match.group(0) if match else None

    def get_response(self, user_input):
        intent = self.detect_intent(user_input)
        order_id = self.extract_order_id(user_input)

        if order_id and order_id in self.ORDERS:
            self.current_order = order_id
            order = self.ORDERS[order_id]
            response = f"Order {order_id}: {order['status']} ({order['date']} - {order['location']})"
        elif intent == "order_status" and self.current_order:
            order = self.ORDERS[self.current_order]
            response = f"Your recent order {self.current_order}: {order['status']} ({order['location']})"
        else:
            response = self.RESPONSES.get(intent, self.RESPONSES["unknown"])
            if isinstance(response, list):
                response = response[0]  # Pick first for consistency

        # Greeting personalization
        if intent == "greeting":
            self.user_name = user_input.split()[-1].title() if len(user_input.split()) > 1 else "Customer"
            response = response.replace("!", f" {self.user_name}!")

        timestamp = datetime.now().strftime("%H:%M")
        full_response = f"[{timestamp}] {response}"

        self.chat_history.append({"user": user_input, "bot": full_response})
        return full_response

def run_perfect_bot():
    bot = PerfectCustomerBot()
    print("ðŸ¤– BlueCart PERFECT Support Bot Started! (No HF weirdness)")
    print("ðŸ’¬ Demo questions: 'track ORD1002', 'return policy?', 'shipping?'\n")

    while True:
        user_input = input("ðŸ’¬ You: ").strip()

        if user_input.lower() in ['quit', 'exit', 'bye']:
            print("ðŸ¤– Bot: Thanks for chatting! Have a great day! ðŸ‘‹")
            break

        if not user_input:
            print("ðŸ¤– Bot: Please type something!")
            continue

        print("ðŸ¤– Bot: ", end="", flush=True)
        time.sleep(1)  # Realistic delay

        response = bot.get_response(user_input)
        print(response)
        print("-" * 60)

    # Show final history
    print("\nðŸ“œ Final Chat Summary:")
    for i, chat in enumerate(bot.chat_history[-3:], 1):
        print(f"{i}. You: {chat['user']}")
        print(f"   Bot: {chat['bot']}")

# ðŸ”¥ RUN THIS - PERFECT RESPONSES GUARANTEED
run_perfect_bot()
```
